```python
df = pd.read_csv('out/results/Resultsnewdata.csv')
df2 = pd.read_csv('out/results/ResultsSVR.csv')
df3 = pd.read_csv('out/results/ResultsLR.csv')
# concat both dfs
df = pd.concat([df, df2, df3])
# create new column named params, which contains only the params of the model
df['params'] = df['name'].apply(lambda x: ''.join(x.split('-')[1:])[:-10])
# transform name column, so that it contains only the name of the model
df['fullname'] = df['name']
df['name'] = df['name'].apply(lambda x: x.split('-')[0])
display(df.head())
# scatter f1 scores and  roc_auc_score, color dots based on name
fig, ax = plt.subplots()
for name, group in df.groupby('name'):
      ax.scatter(group['f1_score'], group['roc_auc_score'], label=name)
# show legend
ax.legend()
# set labels
ax.set_xlabel('f1_score')
ax.set_ylabel('roc_auc_score')
# set title
ax.set_title('f1_score vs roc_auc_score')
# show plot
plt.show()
```

```python
# create a new plot, same as above but only with logisticregression and svr and dont group them, each point will have its own color
fig, ax = plt.subplots()
# dont group by name, each row will have own color
for name, group in df[((df['name'] == 'LogisticRegression') | (df['name'] == 'SVR')) & (df['roc_auc_score'] > 0.89)].groupby('params'):
      # if name contains word gamma
        if 'gamma' in name:
            ax.scatter(group['f1_score'], group['roc_auc_score'], label=name, marker='x')
        else:
            ax.scatter(group['f1_score'], group['roc_auc_score'], label=name, marker='o')
# print legend

# set labels
# change legend location
ax.legend(loc='upper left')
# move legend outside of plot and make it smaller
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0., fontsize='small')
ax.set_xlabel('f1_score')
ax.set_ylabel('roc_auc_score')
# set title
ax.set_title('f1_score vs roc_auc_score')
# show plot
plt.show()

```

```python
df = df.sort_values(by=['roc_auc_score'], ascending=False)
# Print top 10
print(f'ROC AUC Score')
display(df[['roc_auc_score', 'fullname']].head(5))
```
